```python
import pandas as pd
import ast
```

```python
#leemos los archivos y los guardamos en un dataframe

pelis1 = pd.read_csv('movies_dataset.csv')
pelis2 = pd.read_excel('movies_dataset.xlsx')
pelis3 = pd.read_excel('movies_dataset(1).xlsx')
credits = pd.read_csv('credits.csv')

#concatenamos los dataframes
peliculasAll = pd.concat([pelis1, pelis2, pelis3])
#concatenamos los dataframes de peliculas con el de creditos por la columna 'id'
peliculas = pd.merge(peliculasAll, credits, on='id')



```

```python
#Los valores nulos de los campos revenue, budget deben ser rellenados por el número 0
peliculas['revenue'] = peliculas['revenue'].fillna(0)
peliculas['budget'] = peliculas['budget'].fillna(0)

peliculas
```

```python
#Los valores nulos del campo release date deben eliminarse.
peliculas = peliculas.dropna(subset=['release_date'])
peliculas
```

```python
#Eliminar las columnas que no serán utilizadas, video,imdb_id,adult,original_title,poster_path y homepage
peliculas = peliculas.drop(['video','imdb_id','adult','original_title','poster_path','homepage'], axis=1)
peliculas
```

```python
#De haber fechas, deberán tener el formato AAAA-mm-dd, además deberán crear la columna release_year donde extraerán el año de la fecha de estreno.
peliculas['release_date'] = pd.to_datetime(peliculas['release_date'], errors='coerce')
peliculas['release_year'] = peliculas['release_date'].dt.year
peliculas
```

```python
#vamos a eliminar las remplazar los valores string terminados con jpg por 0
def convert_to_int(value):
    try:
        return int(value)
    except ValueError:
        return 0

peliculas['budget'] = peliculas['budget'].astype(str)
peliculas['budget'] = peliculas['budget'].str.replace(r'\.jpg', '0')
peliculas['budget'] = peliculas['budget'].apply(convert_to_int)

#Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, dividiendo estas dos últimas revenue / budget, cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0
peliculas['return'] = peliculas['revenue'] / peliculas['budget']
peliculas
```

```python
#desanidamos los datos de la columna 'spoken_languages' por los que sigan despues de 'name


peliculas['spoken_languages'] = peliculas['spoken_languages'].apply(lambda x: [item['name'] for item in ast.literal_eval(x)] if isinstance(x, str) else [])
peliculas
```

```python
#desanidamos los datos de belogns_to_collection
def fetch_names(obj):
    names = []
    if isinstance(obj, str) and '{' in obj:
        dic = ast.literal_eval(obj)
        return dic['name']
peliculas['belongs_to_collection'] = peliculas['belongs_to_collection'].apply(fetch_names)
peliculas
```

```python
#columna belogns_to_collection a tipo lista
peliculas['belongs_to_collection'] = peliculas['belongs_to_collection'].apply(lambda x: [x] if isinstance(x, str) else [])
peliculas
```

```python
peliculas.to_csv('movies_dataset_clean.csv', index=False)
```
